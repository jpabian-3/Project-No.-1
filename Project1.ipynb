```python
import pandas as pd
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')
#Request 1
```

```python
air_data_grouped = air_data.groupby('sensor.name')
air_data_pm25 = air_data_grouped.agg(max = ('pm2.5_atm', 'max'),
                                                    min = ('pm2.5_atm', 'min'),
                                                    mean = ('pm2.5_atm', 'mean'),
                                                    median = ('pm2.5_atm', 'median'))
print('Statistics for highest mean PM 2.5 Values')
top5_mean_pm25 = (
    air_data_pm25.sort_values(by = 'mean', ascending = False) #sort_values by any specified column
)

top5_mean_pm25.head(5)
```

```python
print('Statistics for highest median PM 2.5 Values')
top5_median_pm25 = (
    air_data_pm25.sort_values(by = 'median', ascending = False) #sort_values by any specified column
)

top5_median_pm25.head(5)
```

```python
#Request 3
def assign_temp(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot'

air_data['temperature_level'] = air_data['temperature'].apply(assign_temp) #the input to my assign temp will be the data column temperature
air_data.head(5)
```

```python
def assign_humid(humidity):
    if humidity < 50:
        return 'Low Humidity'
    elif 50 <= humidity <= 80:
        return 'High Humidity'
    else:
        return 'Very High Humidity'

air_data['humidity_level'] = air_data['humidity'].apply(assign_humid) #the input to my assign temp will be the data column temperature
air_data.head(5)
```

```python
#Request 4
def aqi_index(quality):
    if quality <= 50:
        return 'Good'
    elif 51 <= quality <= 100:
        return 'Moderate'
    elif 101 <= quality <= 150:
        return 'Unhealthy for Sensitive Groups'
    elif 151 <= quality <= 200:
        return 'Unhealthy'
    elif 201 <= quality <= 300:
        return 'Very Unhealthy'
    else:
        return 'Hazardous'

def assign_temp(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot'

air_data['temperature_level'] = air_data['temperature'].apply(assign_temp)
air_data['aqi_level'] = air_data['pm2.5_atm'].apply(aqi_index) 
air_data_unhealthy_sg = air_data[air_data['aqi_level'] == 'Unhealthy for Sensitive Groups'
]
air_data_unhealthy_sg
```
